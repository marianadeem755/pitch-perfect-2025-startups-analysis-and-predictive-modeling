## **Impute Missing values using Iterative Imputer and Simple Imputer**

```python
# Iterative Imputer for high missing
iter_imputer = IterativeImputer(random_state=42)
df[['Last_Funding_Amount_USD_Millions']] = iter_imputer.fit_transform(df[['Last_Funding_Amount_USD_Millions']])

# Simple Imputer (median) for moderate missing
simple_imputer = SimpleImputer(strategy='median')
df[['Number_of_Employees']] = simple_imputer.fit_transform(df[['Number_of_Employees']])
```

## **Impute Missing values using Mean, Median and Mode**

```python
# Categorical columns (mode)
cat_cols = ['Funding_Stage', 'Website', 'LinkedIn_Profile', 'Twitter_Handle']
for col in cat_cols:
    mode_val = df[col].mode()[0]
    df[col].fillna(mode_val, inplace=True)

# Numerical columns (median or mean)
# Median
median_cols = ['Last_Funding_Amount_USD_Millions', 'Number_of_Employees']
for col in median_cols:
    median_val = df[col].median()
    df[col].fillna(median_val, inplace=True)

# Mean
mean_cols = ['Market_Size_Billion_USD']
for col in mean_cols:
    mean_val = df[col].mean()
    df[col].fillna(mean_val, inplace=True)
```

## **Lets check the Correlation between Employees and Funding, Funding Efficiency per Employee & Market Funding**

```python
# Correlation between Employees and Funding
corr = np.corrcoef(df["Number_of_Employees"], df["Last_Funding_Amount_USD_Millions"])[0,1]
print("Correlation Analysis")
print("Correlation between employees and funding:", round(corr, 3))
if corr > 0.5:
    print("Strong positive relation: More employees generally means higher funding")
elif corr < -0.5:
    print("Strong negative relation: More employees tends to mean lower funding")
else:
    print("Weak/No clear relation between employees and funding")
print("==========================================================================")

# Funding Efficiency Funding per Employee 
df["Funding_per_Employee"] = df["Last_Funding_Amount_USD_Millions"] / df["Number_of_Employees"].replace(0, np.nan)
efficiency_stats = df.groupby("Industry")["Funding_per_Employee"].mean().sort_values(ascending=False)
print("Funding Efficiency by Industry")
print("Highest capital per employee:", efficiency_stats.idxmax(), "is", round(efficiency_stats.max(),2))
print("Lowest capital per employee:", efficiency_stats.idxmin(), "is", round(efficiency_stats.min(),2))
print("==========================================================================")

# Market Funding (Funding / Market Size)
df["Funding_to_Market_Ratio"] = df["Last_Funding_Amount_USD_Millions"] / df["Market_Size_Billion_USD"].replace(0, np.nan)
ratio_stats = df.groupby("Industry")["Funding_to_Market_Ratio"].mean().sort_values(ascending=False)
print("Funding relative to Market Size")
print("Industry most aggressive vs market size:", ratio_stats.idxmax(), "is", round(ratio_stats.max(),2))
print("Industry least aggressive vs market size:", ratio_stats.idxmin(), "is", round(ratio_stats.min(),2))
```

## **Define Features and Target Variables and Handle Class Imbalance**

```python
# Define Features and Target
X = df.drop(columns=['Funding_Stage'])  # Features
y = df['Funding_Stage']                 # Target

# Encode categorical variables if needed
for col in X.select_dtypes(include='object').columns:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col].astype(str))

# Handle Class Imbalance by Upsampling
df_full = pd.concat([X, y], axis=1)
max_count = y.value_counts().max()

df_list = []
for cls in y.unique():
    df_cls = df_full[df_full['Funding_Stage'] == cls]
    df_cls_upsampled = df_cls.sample(max_count, replace=True, random_state=42)
    df_list.append(df_cls_upsampled)

df_resampled = pd.concat(df_list)

X_res = df_resampled.drop(columns=['Funding_Stage'])
y_res = df_resampled['Funding_Stage']
```

## **Train and Evaluate KNN, Decision Tree & Random Forest**

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.2, random_state=42, stratify=y_res)

# Initialize models
models = {
    'KNN': KNeighborsClassifier(n_neighbors=5, weights='distance'),
    'Decision_Tree': DecisionTreeClassifier(random_state=42),
    'Random_Forest': RandomForestClassifier(n_estimators=100, random_state=42)
}

# Dictionary to store performance metrics
performance = {}

# Train and evaluate each model
for name, model in models.items():
    # For KNN, apply scaling + PCA
    if name == 'KNN':
        scaler = StandardScaler()
        X_train_scaled = scaler.fit_transform(X_train)
        X_test_scaled = scaler.transform(X_test)

        pca = PCA(n_components=0.95, random_state=42)
        X_train_pca = pca.fit_transform(X_train_scaled)
        X_test_pca = pca.transform(X_test_scaled)

        model.fit(X_train_pca, y_train)
        y_pred = model.predict(X_test_pca)
    else:
        # Decision Tree / Random Forest
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
    
    # Calculate metrics
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, average='macro')
    rec = recall_score(y_test, y_pred, average='macro')
    f1 = f1_score(y_test, y_pred, average='macro')
    
    performance[name] = {'Accuracy': acc, 'Precision': prec, 'Recall': rec, 'F1-score': f1}
    
    # Print metrics
    print(f"\n================={name} =====================")
    print("Accuracy:", acc)
    print("Precision (macro):", prec)
    print("Recall (macro):", rec)
    print("F1-score (macro):", f1)
    print("\nClassification Report:\n", classification_report(y_test, y_pred))
    print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))

# Create DataFrame for heatmap
perf_df = pd.DataFrame(performance).T  # transpose to have models as rows
plt.figure(figsize=(8, 6))
sns.heatmap(perf_df, annot=True, cmap='viridis', fmt=".3f")
plt.title("Model Performance Comparison")
plt.show()
```

## **Hyperparameter Tuning, Model Evaluation, & Comparison, Stacking & Ensembling**

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.2, random_state=42, stratify=y_res)

# Feature Scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# KNN Hyperparameter Tuning
knn = KNeighborsClassifier()

param_grid = {
    'n_neighbors': [3, 5, 7, 9, 11],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan', 'minkowski']
}

cv_strategy = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

grid_search = GridSearchCV(knn, param_grid, cv=cv_strategy, scoring='f1_macro', n_jobs=1)
grid_search.fit(X_train_scaled, y_train)

best_knn = grid_search.best_estimator_
print("Best KNN parameters:", grid_search.best_params_)

# Other Models
dt = DecisionTreeClassifier(random_state=42)
rf = RandomForestClassifier(n_estimators=100, random_state=42)

# Voting Ensemble
voting_clf = VotingClassifier(
    estimators=[('rf', rf), ('dt', dt), ('knn', best_knn)],
    voting='soft',
    weights=[3, 2, 1] 
)
voting_clf.fit(X_train_scaled, y_train)

# Stacking Ensemble
stacking_clf = StackingClassifier(
    estimators=[('rf', rf), ('dt', dt), ('knn', best_knn)],
    final_estimator=LogisticRegression(max_iter=1000, random_state=42),
    passthrough=True,
    n_jobs=-1
)
stacking_clf.fit(X_train_scaled, y_train)

# Evaluation Function
def evaluate_model(name, model, X_test, y_test):
    y_pred = model.predict(X_test)
    print(f"\n=== {name} ===")
    print("Accuracy:", accuracy_score(y_test, y_pred))
    print("Precision (macro):", precision_score(y_test, y_pred, average='macro'))
    print("Recall (macro):", recall_score(y_test, y_pred, average='macro'))
    print("F1-score (macro):", f1_score(y_test, y_pred, average='macro'))
    print("\nClassification Report:\n", classification_report(y_test, y_pred))
    print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))

# Run Evaluations
evaluate_model("Best KNN", best_knn, X_test_scaled, y_test)
evaluate_model("Decision Tree", dt.fit(X_train_scaled, y_train), X_test_scaled, y_test)
evaluate_model("Random Forest", rf.fit(X_train_scaled, y_train), X_test_scaled, y_test)
evaluate_model("Voting Ensemble", voting_clf, X_test_scaled, y_test)
evaluate_model("Stacking Ensemble", stacking_clf, X_test_scaled, y_test)

# Compare Results in Table
models = {
    "KNN": best_knn,
    "Decision Tree": dt,
    "Random Forest": rf,
    "Voting Ensemble": voting_clf,
    "Stacking Ensemble": stacking_clf
}

results = {}
for name, model in models.items():
    y_pred = model.predict(X_test_scaled)
    results[name] = {
        "Accuracy": accuracy_score(y_test, y_pred),
        "Precision": precision_score(y_test, y_pred, average='macro'),
        "Recall": recall_score(y_test, y_pred, average='macro'),
        "F1": f1_score(y_test, y_pred, average='macro')
    }

results_df = pd.DataFrame(results).T
print("\n========================= Performance Comparison ==========================\n")
print(results_df)

# Heatmap for visualization
plt.figure(figsize=(8,5))
sns.heatmap(results_df, annot=True, fmt=".3f", cmap="viridis")
plt.title("Model Performance Comparison", fontsize=14)
plt.show()
```

```python

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.2, random_state=42, stratify=y_res)

classes = np.unique(y_test)


# Scale and PCA for KNN
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

pca = PCA(n_components=0.95, random_state=42)
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)

# Default Models
default_models = {
    'KNN': KNeighborsClassifier(n_neighbors=5, weights='distance'),
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42)
}

default_results = {}
for name, model in default_models.items():
    if name == 'KNN':
        model.fit(X_train_pca, y_train)
        y_pred = model.predict(X_test_pca)
        y_prob = model.predict_proba(X_test_pca)
    else:
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        y_prob = model.predict_proba(X_test)
    
    y_test_bin = label_binarize(y_test, classes=classes)
    roc_auc_macro = roc_auc_score(y_test_bin, y_prob, average='macro', multi_class='ovr')

    default_results[name] = {
        'Accuracy': accuracy_score(y_test, y_pred),
        'Precision': precision_score(y_test, y_pred, average='macro'),
        'Recall': recall_score(y_test, y_pred, average='macro'),
        'F1': f1_score(y_test, y_pred, average='macro'),
        'ROC_AUC': roc_auc_macro
    }

# Tuned Models 
# RandomizedSearchCV
cv_strategy = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# KNN tuned
knn_param_dist = {
    'n_neighbors': randint(3, 20),
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan', 'minkowski']
}
knn_search = RandomizedSearchCV(
    KNeighborsClassifier(),
    knn_param_dist,
    n_iter=30,
    cv=cv_strategy,
    scoring='f1_macro',
    n_jobs=-1,
    random_state=42
)
knn_search.fit(X_train_pca, y_train)
best_knn = knn_search.best_estimator_

# Decision Tree tuned
dt_param_dist = {
    'max_depth': randint(3, 20),
    'min_samples_split': randint(2, 20),
    'min_samples_leaf': randint(1, 10),
    'criterion': ['gini', 'entropy']
}
best_dt = RandomizedSearchCV(
    DecisionTreeClassifier(random_state=42),
    dt_param_dist,
    n_iter=30,
    cv=cv_strategy,
    scoring='f1_macro',
    n_jobs=-1,
    random_state=42
).fit(X_train, y_train).best_estimator_

# Random Forest tuned
rf_param_dist = {
    'n_estimators': randint(50, 300),
    'max_depth': randint(3, 20),
    'min_samples_split': randint(2, 20),
    'min_samples_leaf': randint(1, 10),
    'max_features': ['auto', 'sqrt', 'log2']
}
best_rf = RandomizedSearchCV(
    RandomForestClassifier(random_state=42),
    rf_param_dist,
    n_iter=30,
    cv=cv_strategy,
    scoring='f1_macro',
    n_jobs=-1,
    random_state=42
).fit(X_train, y_train).best_estimator_

tuned_models = {
    'KNN': best_knn,
    'Decision Tree': best_dt,
    'Random Forest': best_rf
}

tuned_results = {}
for name, model in tuned_models.items():
    if name == 'KNN':
        y_pred = model.predict(X_test_pca)
        y_prob = model.predict_proba(X_test_pca)
    else:
        y_pred = model.predict(X_test)
        y_prob = model.predict_proba(X_test)
    
    y_test_bin = label_binarize(y_test, classes=classes)
    roc_auc_macro = roc_auc_score(y_test_bin, y_prob, average='macro', multi_class='ovr')

    tuned_results[name] = {
        'Accuracy': accuracy_score(y_test, y_pred),
        'Precision': precision_score(y_test, y_pred, average='macro'),
        'Recall': recall_score(y_test, y_pred, average='macro'),
        'F1': f1_score(y_test, y_pred, average='macro'),
        'ROC_AUC': roc_auc_macro
    }

# Combine Results
metrics = ['Accuracy', 'Precision', 'Recall', 'F1', 'ROC_AUC']
comparison_data = {}
for metric in metrics:
    comparison_data[metric] = []
    for name in default_results.keys():
        comparison_data[metric].append(default_results[name][metric])
        comparison_data[metric].append(tuned_results[name][metric])

comparison_df = pd.DataFrame(
    comparison_data,
    index=[f"{name} Default" for name in default_results.keys()] + [f"{name} Tuned" for name in tuned_results.keys()]
)

print("\n=== Default vs Tuned Models Comparison ===\n")
print(comparison_df)

# Heatmap
plt.figure(figsize=(10,6))
sns.heatmap(comparison_df, annot=True, fmt=".3f", cmap="viridis")
plt.title("Default vs Tuned Models - All Metrics", fontsize=14)
plt.show()
```
